```python
with open('E:\\VIT UNIVERSITY  ACADEMICS\\vit assignment\\WINTER SEMESTER\\Nosql Database J component Review\\Data Collection for Cardiovascular Disease\\Data Collection for Heart Disease.csv',encoding="utf-8") as cardiovascular_data:
    csv_reader = csv.DictReader(cardiovascular_data)
    with open('new_cardiovascular_disease_dataset.csv','w') as new_cardiovascular_data:
        fieldnames = ['Timestamp', 'Username', 'Name ',
       'Your age? please select appropriate age group', 'Your Gender?',
       'did u have the Heart disease ?', 'BP Level?', 'Do you have sugar?',
       'What is the sugar level?', 'What is the Cholestrol level?',
       'What is the Glucose level?', 'Have you ever done smoking?',
       'How often did you drink liquor?', 'What is your Height ?',
       'What is your Weight?']
        csv_writer = csv.DictWriter(new_cardiovascular_data,fieldnames=fieldnames,delimiter=',')
        csv_writer.writeheader()
        for line in csv_reader:
            csv_writer.writerow(line)
```

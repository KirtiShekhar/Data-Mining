```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = "original facial image.png"    
img = cv2.imread(image_path,1);
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB);
```

```python
# reading or displaying the image using matplotlib
plt.imshow(img)
plt.title("original image")
plt.show()
```

```python
#implementing masking using sharpening and gaussian blur
print('implementing masking using sharpening and gaussian blur')
```

```python
#defining the matrix for sharpening
print('defining the matrix for sharpening')
k_sharp = np.array(([0,-1,0],
                    [-1,5,-1],
                    [0,-1,0]),np.float32)
```

```python
print('sharpening matrix')
print(k_sharp)
```

```python
output_sharp = cv2.filter2D(img,-1,k_sharp)
```

```python
#defining the matrix for gaussian blur
print('defining the matrix for gaussian blur')
k_gaussian_blur = np.array(([1,2,1],
                    [2,4,2],
                    [1,8,9]),np.float32)/16
```

```python
print('gaussian blur matrix')
print(k_gaussian_blur)
```

```python
output_gaussian_blur = cv2.filter2D(img,-1,k_gaussian_blur)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(output_sharp)
plt.title('Filtered Sharpen Image')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(output_gaussian_blur)
plt.title('Filtered Gaussian blur Image')
```

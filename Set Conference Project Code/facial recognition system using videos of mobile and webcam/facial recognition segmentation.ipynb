```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = "original facial image.png"    
img = cv2.imread(image_path,1);
```

```python
# reading or displaying the image using matplotlib
plt.imshow(img)
plt.title("original image")
plt.show()
```

```python
# applying segmentation using thresholding 
th = 127
max_val = 255
ret,output_img1 = cv2.threshold(img,th,max_val,cv2.THRESH_BINARY)
ret,output_img2 = cv2.threshold(img,th,max_val,cv2.THRESH_BINARY_INV)
ret,output_img3 = cv2.threshold(img,th,max_val,cv2.THRESH_TOZERO)
ret,output_img4 = cv2.threshold(img,th,max_val,cv2.THRESH_TOZERO_INV)
ret,output_img5 = cv2.threshold(img,th,max_val,cv2.THRESH_TRUNC)
```

```python
# array of output and titles
output = [img,output_img1,output_img2,output_img3,output_img4,output_img5]
titles = ['original_image','binary_image','binary_invereted_image','zero_image','zero_inverted_image','truncated_image']
```

```python
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(output[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
